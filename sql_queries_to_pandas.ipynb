```
### select * from data where gender='M'
df[df.Gender=='M']
```

```
### select * from Data where Gender='F' and Education_Level='High School'
df[(df.Gender=='F') & (df.Education_Level=='High School')]
```

```
### select count(*) from Data where Gender='F' and Education_Level='High School'
df[(df.Gender=='F') & (df.Education_Level=='High School')].count()
```

```
### select customer_age,Gender,Education_Level from Data
df[['Customer_Age','Gender','Education_Level']]
```

```
### select Gender,Education_Level,count(*) from Data group by Gender,Education_Level order by size desc

### size includes NaN values while count doesnt not include Nan
df.groupby(['Gender','Education_Level'],as_index=False).size().sort_values(by='size',ascending=False).rename(columns={'size':'count'})
```

```
### select Card_Category, count(clientnum) from Data group by Card_Category
df.groupby(['Card_Category'],as_index=False).agg({'CLIENTNUM':'count'})
```

```
### select Card_Category, count(distinct clientnum) from Data group by Card_Category
df.groupby(['Card_Category'],as_index=False).agg({'CLIENTNUM':'nunique'})
```

```
### select Gender, Education_Level, sum(Trans_Amt) AS 'Total Amount',count(distinct clientnum) AS 'item_count' from Data group by 1,2 order by sum(Trans_Amt) desc
### use \ to wrap pandas commands in multiple lines
df.groupby(['Gender','Education_Level'],as_index=False).agg({'Total_Trans_Amt':'sum','CLIENTNUM':'nunique'})\
.rename(columns={'Total_Trans_Amt':'Total Amount','CLIENTNUM':'item_count'})\
.sort_values(by=['Total Amount'],ascending=False)
```

```
## select CLIENTNUM,Education_Level,sum(Trans_Amt) AS 'Total Amount' from Data where Marital_Status='Single' and Gender='F' and Months_on_book>=6 
## group by 1,2 having(sum(Trans_Amt)>10000) order by sum(Trans_Amt) desc

df2=df[(df.Gender=='F') & (df.Marital_Status=='Single') &(df.Months_on_book>=6)].groupby(['CLIENTNUM','Education_Level'],as_index=False)\
.agg({'Total_Trans_Amt':'sum'}).rename(columns={'Total_Trans_Amt':'Total Amount'}).sort_values(by=['Total Amount'],ascending=[False]).reset_index(drop=True)

df2[df2['Total Amount']>10000]
```

```
## select * from Data where Education in ('Graduate','High School	')
df[df.Education_Level.isin(['Graduate','High School'])]
```
